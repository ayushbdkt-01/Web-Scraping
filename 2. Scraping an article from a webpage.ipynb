```python
import requests     # library for requesting server
from bs4 import BeautifulSoup    # library for web scraping
```

```python
res=requests.get('https://quotes.toscrape.com/')
soup=BeautifulSoup(res.text,'html.parser')
soup      # storing the code of the webpage in the variable 'soup'
```

```python
article=soup.find('span',class_='text').text    # scraps a single article using a class
# find is used to scrape a single article from the webpage
```

```python
print(article)
```

```python
soup.find_all('span',class_='text')    # scraps all the articles in the webpage
```

```python
for articles in soup.find_all('span',class_='text'):
    print(articles.text[1:-1])           # printing all the articles from the webpage
    print()
```
