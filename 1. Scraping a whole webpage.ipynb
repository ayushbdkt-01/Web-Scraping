```python
import requests      # used to send HTTP requests to the server
```

# scraping a whole web-page

```python
link='https://quotes.toscrape.com/'
```

```python
link
```

```python
response=requests.get(link)    # storing the response from the server
```

```python
response
```

```python
print(response.text)      # scraps the whole code of the webpage
```

```python
html=response.text
html=html.encode('utf-8')     # changing the encoding of the file
html=str(html)   # converting to string
```

```python
fd=open('index.html','w')
fd.write(html)
fd.close()
```
